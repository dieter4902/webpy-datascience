```python
from keras.datasets import mnist
from keras.datasets import fashion_mnist
from keras.layers import Dense
from keras.models import Sequential
from keras.utils.np_utils import to_categorical
from matplotlib import pyplot as plt
```

```python
# Setup train and test splits
(x_train, y_train), (x_test, y_test) = fashion_mnist.load_data()
```

```python
# Inspect train data
print("Training data shape: ", x_train.shape)  # (60000, 28, 28) -- 60000 images, each 28x28 pixels

# Show the first few train images as a plot
for i in range(9):
    plt.subplot(330 + 1 + i)  # Define subplot
    plt.imshow(x_train[i], cmap=plt.get_cmap('gray'))  # Plot raw pixel data
plt.show()
```

```python
# Inspect test data
print("Test data shape: ", x_test.shape)  # (10000, 28, 28) -- 10000 images, each 28x28 pixels

# Show the first few test images as a plot
for i in range(9):
    plt.subplot(330 + 1 + i)  # Define subplot
    plt.imshow(x_test[i], cmap=plt.get_cmap('gray'))  # Plot raw pixel data
plt.show()
```

```python
# Inspect labels
print("Training labels shape: ", y_train.shape)  # (60000, 28, 28) -- 60000 numbers from the range 0, 1, ... 9
print("Test labels shape: ", y_test.shape)  # (10000, 28, 28) -- 10000 numbers from the range 0, 1, ... 9
print("First 10 training labels: ", y_train[:10])
print("First 10 test labels: ", y_test[:10])
```

```python
# Flatten the images
image_vector_size = 28*28
x_train = x_train.reshape(x_train.shape[0], image_vector_size)
x_test = x_test.reshape(x_test.shape[0], image_vector_size)
```

```python
# Re-inspect data shapes
# 28 * 28 = 784
print("Training data shape: ", x_train.shape)  # (60000, 784) -- 60000 images, each a flat series of 784 pixels
print("Test data shape: ", x_test.shape)  # (10000, 784) -- 10000 images, a flat series of 784 pixels
```

```python
# Convert to "one-hot" vectors using the to_categorical function
num_classes = 10
y_train = to_categorical(y_train, num_classes)
y_test = to_categorical(y_test, num_classes)
```

```python
# Re-inspect train labels
print("First 10 training labels as one-hot encoded vectors:\n", y_train[:10])
```

```python
# Re-inspect test labels
print("First 10 test labels as one-hot encoded vectors:\n", y_test[:10])
```

```python
# Define a neural network model
image_size = 784
num_classes = 10

model = Sequential()  # Documentation: https://keras.io/models/sequential/

# The input layer requires the special input_shape parameter which should match
# the shape of our training data.
model.add(Dense(units=32, activation='sigmoid', input_shape=(image_size,)))  # Dense = fully connected layers
model.add(Dense(units=num_classes, activation='softmax'))
model.compile(optimizer="sgd", loss='categorical_crossentropy', metrics=['accuracy'])
model.summary()

# Define a neural network model
model1 = Sequential()
model1.add(Dense(units=32, activation='swish', input_shape=(image_size,)))
model1.add(Dense(units=num_classes, activation='elu'))
model1.add(Dense(units=32, activation='softmax'))
model1.add(Dense(units=num_classes, activation='sigmoid'))
model1.add(Dense(units=32, activation='sigmoid'))
model1.add(Dense(units=num_classes, activation='softmax'))
model1.compile(optimizer="sgd", loss='categorical_crossentropy', metrics=['accuracy'])
model1.summary()
```

```python
# Train the model and keep track of progress
history = model.fit(x_train, y_train, batch_size=128, epochs=200, verbose=False, validation_split=.1)
```

```python
history1 = model1.fit(x_train, y_train, batch_size=128, epochs=200, verbose=False, validation_split=.1)
```

```python
# Evaluate the model
loss, accuracy  = model.evaluate(x_test, y_test, verbose=False)
```

```python
loss1, accuracy1  = model1.evaluate(x_test, y_test, verbose=False)
```

```python
# Display the results
print('test with 50 epochs')
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.plot(history1.history['accuracy'])
plt.plot(history1.history['val_accuracy'])
plt.title('Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend(['Train', 'Test','Train1', 'Test1'], loc='best')
plt.show()

plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.plot(history1.history['loss'])
plt.plot(history1.history['val_loss'])
plt.title('Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend(['Train', 'Test','Train1', 'Test1'], loc='best')
plt.show()

print('sigmoid -> softmax test loss:', loss)
print('sigmoid -> softmax test accuracy:', accuracy)
print()
print('swish -> elu -> softmax -> sigmoid -> softmax -> sigmoid test loss:', loss1)
print('swish -> elu -> softmax -> sigmoid -> softmax -> sigmoid test accuracy:', accuracy1)
```

during the first 25 epochs it looks like the improvement of the second model is improving exponentially compared to the logarythmic growth of the first model, but after that the numbers just jump around and loose their reliability compared to the first model
